```python
metadata=MetaData()
```

```python
print(disk_engine.table_names())
```

```python
df_event_level_opt_new=Table('event_level', metadata, autoload=True, autoload_with=disk_engine)
```

```python
print(df_event_level_opt_new.columns.keys())
```

```python
df_transaction_new=Table('trans', metadata, autoload=True, autoload_with=disk_engine)
```

```python
print(df_transaction_new.columns.keys())
```

```python
elpt=select([df_event_level_opt_new, df_transaction_new])
```

```python
elpt=elpt.select_from(
df_event_level_opt_new.join(df_transaction_new,df_event_level_opt_new.columns.EquipmentID==df_transaction_new.columns.TRNUNT))
```
